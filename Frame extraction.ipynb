```python
import cv2
import os
from pathlib import Path

# Define the path to the folder containing the video files
videos_folder_path = 'C:\\Users\\iGauze\\Desktop\\Videos 4'  # Adjust this path to your videos folder
video_extensions = ['*.mp4', '*.mov']  # List of video file extensions to look for

# Create the "images" folder on your desktop if it doesn't exist
desktop_path = os.path.join(os.path.expanduser('~'), 'Desktop')
images_folder = os.path.join(desktop_path, 'images 4')
os.makedirs(images_folder, exist_ok=True)

# Set the frame rate and time interval for frame extraction
fps = 11.04  # Approximate frames per second of the videos
time_interval = 3  # Seconds

# Function to process a single video file
def process_video(video_file):
    cap = cv2.VideoCapture(video_file)
    if not cap.isOpened():
        print(f"Error: Could not open video file {video_file}")
        return

    # Calculate frame interval based on FPS and desired time interval
    frame_interval = int(fps * time_interval)
    frame_count = 0

    while True:
        ret, frame = cap.read()

        if not ret:
            break  # End of video

        frame_count += 1

        # Extract a frame every 'frame_interval' frames (approximately every 3 seconds)
        if frame_count % frame_interval == 0:
            # Save the frame to the common "images" folder
            frame_filename = os.path.join(images_folder, f'{Path(video_file).stem}_frame_{frame_count}.jpg')
            cv2.imwrite(frame_filename, frame)
            #print(f'Saved frame {frame_count} as {frame_filename}')

    # Release the video capture object
    cap.release()

# Main loop to iterate through each video file in the folder
for extension in video_extensions:
    for video_file in Path(videos_folder_path).glob(extension):
        print(f"Processing video {video_file}...")
        process_video(str(video_file))

print("All videos processed.")
```
